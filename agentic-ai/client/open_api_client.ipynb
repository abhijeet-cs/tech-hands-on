```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown, display


load_dotenv(override=True)
```

```python
api_key = os.getenv("GOOGLE_API_KEY")
if not api_key:
    raise ValueError("GOOGLE_API_KEY not set")
```

```python
client = OpenAI(
    api_key=api_key,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)
```

```python
messages = [
    {"role": "user", "content": "Hello! Tell me something interesting."}
]

response = client.chat.completions.create(
    model="gemini-3.1-flash-lite-preview",
    messages=messages
)

print(response.choices[0].message.content)
```

```python
#Markdown 
display(Markdown(response.choices[0].message.content))
```

```python
# Structured outputs
from pydantic import BaseModel

class Summarize(BaseModel):
    summarized_message: str
    original_message: str
```

```python
messages = [{"role": "system", "content": "You are a summarizer. Please summarize the response in 5 words and responsd wih summarized and original message"}] + [{"role": "user", "content": "Hello! Tell me something interesting"}]
```

```python
response = client.beta.chat.completions.parse(model="gemini-2.5-flash", messages=messages, response_format=Summarize)
print(response.choices[0].message.parsed)
```
